```python
import numpy as np
import matplotlib.pyplot as plt

W = 30
H = 30

img_data = np.zeros((W, H))
```

```python
def line_dda(arr, start, end):
    ret = np.copy(arr)
    
    x0, y0 = start
    x1, y1 = end
    
    dx = x1 - x0
    dy = y1 - y0
    
    r = abs(dx) if abs(dx) > abs(dy) else abs(dy)
    
    xr = dx / r
    yr = dy / r
    
    next_x = x0 
    next_y = y0
    
    ret[next_x, next_y] = 1
    
    while next_x != x1 and next_y != y1:
        next_x = next_x + xr
        next_y = next_y + yr
        ret[round(next_x), round(next_y)] = 1
    
    return ret

def line_bresenham(arr, start, end):
    ret = np.copy(arr)
    
    x0, y0 = start
    x1, y1 = end
    
    dx = x1 - x0
    dy = y1 - y0
    
    m = dy / dx
    
    greater_than_one = False
    d1 = None
    d2 = None
    p = None
    
    if m > 0 and m < 1:
        d1 = 2 * dy
        d2 = 2 * (dx - dy)
        p = d1 - dx
        greater_than_one = False
    elif m > 1:
        d1 = 2 * dx
        d2 = 2 * (dx - dy)
        p = d1 - dy
        greater_than_one = True 
        
    x = x0
    y = y0
    
    ret[x, y] = 1
    
    while x != x1 or y != y1:
        if greater_than_one:
            if p >= 0:
                p = p + d2
                x = x + 1
            else:
                p = p + d1
            y = y + 1
        else:
            if p >= 0:
                p = p - d2
                y = y + 1
            else:
                p = p + d1
            x = x + 1
        ret[x, y] = 1
        
    return ret

def _circle_bresenham_1(arr, pixels, coord):
    x0, y0 = coord
    for x, y in pixels:
        arr[x+x0, y+y0] = 1

def _circle_bresenham_2(arr, pixels, coord):
    x0, y0 = coord
    for x, y in pixels:
        arr[x+x0, -y+y0] = 1
        
def _circle_bresenham_3(arr, pixels, coord):
    x0, y0 = coord
    for x, y in pixels:
        arr[-x+x0, -y+y0] = 1
        
def _circle_bresenham_4(arr, pixels, coord):
    x0, y0 = coord
    for x, y in pixels:
        arr[-x+x0, y+y0] = 1

def circle_bresenham(arr, coord, r):
    ret = np.copy(arr)
    
    x, y = 0, r
    p = 1 - r
    
    pixels = [(x, y)]
    
    while x < y:
        if p < 0:
            x = x + 1
            p = p + 2*x + 1
        else:
            x = x + 1
            y = y - 1
            p = p + 2*x + 1 - 2*y
        
        pixels.append((x, y))
        
    c2 = np.copy(pixels)
    for x, y in c2:
        pixels.append((y, x))
        
    _circle_bresenham_1(ret, pixels, coord)
    _circle_bresenham_2(ret, pixels, coord)
    _circle_bresenham_3(ret, pixels, coord)
    _circle_bresenham_4(ret, pixels, coord)
   
    return ret
```

```python
circle_img = circle_bresenham(img_data, (10, 10), 8)
plt.imshow(circle_img)
```

```python
img = line_dda(img_data, (0, 0), (19, 5))
plt.imshow(img)
```

```python
img_bren = line_bresenham(img_data, (0, 0), (19, 5))
plt.imshow(img_bren)
```
